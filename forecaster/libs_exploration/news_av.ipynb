```python
import requests
import os

import dotenv
```

```python
dotenv.load_dotenv()
```

```python
AV_BASE_URL = "https://www.alphavantage.co/query"
```

```python
AV_API_KEY = os.getenv("ALPHA_VANTAGE_API_KEY")
```

```python
news_sentiment = requests.get(AV_BASE_URL, params={
    "function": "NEWS_SENTIMENT",
    "tickers": "NVDA",
    "topics": "earnings",
    "time_from": "20240501T0000",
    # "time_to": "20240601T0000",
    "sort": "RELEVANCE",
    "apikey": AV_API_KEY,
})

news_sentiment_json = news_sentiment.json()
news_sentiment_json
```

```python
news_sentiment_json["feed"][0]
```

```python
import numpy as np
```

```python
nvidia_ticker_sentiments = [
    float(ticker["ticker_sentiment_score"])
    for post in news_sentiment_json["feed"]
    for ticker in post["ticker_sentiment"]
    if ticker["ticker"] == "NVDA"
]

avg_sentiment = np.mean(nvidia_ticker_sentiments)
std_sentiment = np.std(nvidia_ticker_sentiments)
max_sentiment = np.max(nvidia_ticker_sentiments)
min_sentiment = np.min(nvidia_ticker_sentiments)

avg_sentiment, std_sentiment, max_sentiment, min_sentiment
```
